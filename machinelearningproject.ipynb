```python
import pandas as pd
#for prediction
from sklearn.tree import DecisionTreeClassifier
#for training
from sklearn.model_selection import train_test_split
#for checking the accuracy
from sklearn.metrics import accuracy_score
import joblib
from sklearn import tree

df = pd.read_csv('music.csv')
df
```

```python
input=df.drop(columns=['genre'])
input
```

```python
output=df['genre']
output
```

```python
# for accuracy
# input_train,input_test,output_train,output_test = train_test_split(input, output, test_size=0.2)
```

```python
#for prediction
model=DecisionTreeClassifier()
model.fit(input, output)
tree.export_graphviz(model, out_file='music-recommender.dot',feature_names=['age','gender'],class_names=sorted(output.unique()),
                     label='all',rounded=True,filled=True)
model = joblib.load('music-recommender.joblib')
predictions=model.predict([[21,1]])

#for checking the accuracy
# score=accuracy_score(output_test,predictions)
# score
```
